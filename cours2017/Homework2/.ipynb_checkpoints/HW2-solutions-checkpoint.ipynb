```python
%pylab inline
from scipy.integrate import odeint
```

```python
#Define parameters
R_SS = 2.5*10.**-5 #Subsidence rate in m/yr
P1 = 10.**8 #Period of sea level oscillations in years
P2 = 10.**5
slope = 1./5000.
omega1 = 2.*pi/P1#angular frequency
omega2 = 2.*pi/P2#angular frequency
D = 2.5 *10**5 #m
Psi = R_SS/(slope*D*omega1)
R = omega2/omega1

#Define timestep and number of steps
def dx_dt(xi,ti,Psi, R):
    return -2.*pi*sin(2.*pi*ti) - 2.*pi*R*sin(2.*pi*R*ti) - 2.*pi*Psi*xi     


t = linspace(0,2,100000)
x_ode = odeint(dx_dt, 0., t, args=(Psi,R))
Y_SL = 0.1*(cos(2*pi*t) + cos(2*pi*R*t))

plot(t, x_ode)
plot(t, Y_SL)
legend(['Shoreline','Sea level'])
ylabel('Shoreline/Sea level')
```
