```python
# 1. Derive the coupled ODEs that describe this system
########################
# D_h = small variation in waterlevel
#
#
#
#



# Define parameters
D_h1 = 1 #m
D_h2 = 1 #m
A_R1 = 5000. #m^2
A_R2 = 50. #m^2
L1 = 1000. #m
L2 = 1000.
f1 = 0.1 #unitless
f2 = 0.1
R = 1. #m^3/s
A_c1 = pi*(D_h1**2)/4.
A_c2 = pi*(D_h2**2)/4.

g=9.8 #m^2/s


# Question 1: Coupled equations

def dh_dt(h, tdummy, A_c1,A_c2, D_h1,D_h2, L1, L2, f1, f2, R, A_R1, A_R2):
    h1,h2 = h # what does this part do?
    Cf1 = 1. + f1*L1/D_h1
    Q1 = A_c1*sqrt(2.*g*(h1-h2)/Cf1)
    Cf2 = 1. + f2*L2/D_h2
    Q2 = A_c2*sqrt(2.*g*h2/Cf2)
    
    return ((R - Q1)/A_R1, (Q1 - Q2)/A_R2)



t = linspace(0,10000000,100)
result = odeint(dh_dt,[0.,0.], t, args=(A_c1,A_c2, D_h1,D_h2, L1, L2, f1, f2, R, A_R1, A_R2))

h1_eq = result [-1,0]
h2_eq = result [-1,1]


plot(t,result[:,0])
plot(t,result[:,1])


print 'h1 at equilibrium =', h1_eq
print 'h2 at equilibrium =', h2_eq
```
